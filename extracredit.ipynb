## Problem 1

### 2x + y = 4
### -x - 3y = -7

1) Solution with work

```python
import numpy as np

a = np.array([[2, 1], [-1, -3]])
b = np.array([4, -7])
x = np.linalg.solve(a,b)

print('x = {} and y = {}'.format(x[0],x[1]))
```

2) This represents the point of intersection between the two lines.

## Problem 2

### Next number in pattern 0, 3, 8, 15, 24

Begin by plotting the data:

```python
%matplotlib inline

import matplotlib.pyplot as plt

x = np.array(range(5))
y = np.array([0, 3, 8, 15, 24])

plt.plot(x,y)
```

This can be described by the function:

$2x + x^2 = y$ 

Where x is the index and y is the number in the pattern.

So:

```python
pattern = lambda x: 2*x + x**2
```

So our pattern is:

```python
print(pattern(x))
plt.plot(pattern(x))
```

And the next number is:

```python
pattern(5)
```

And we can visualize this out further:

```python
plt.plot(pattern(np.array(range(1000))))
```

## Problem 3
### How many ways can we arrange [A,B,C,D]

Use itertools for an easy solution:

```python
from itertools import permutations
```

```python
perms = [''.join(p) for p in permutations('ABCD')]
```

```python
print(perms)
len(perms)
```

Or for a coded out solution:

```python
def permute(s):
    out = []
    
    if len(s) == 1:
        out = [s]
    
    else:
        for i, let in enumerate(s):
            for perm in permute(s[:i] + s[i+1:]):
                out += [let + perm]
    
    return out
```

```python
print(permute('ABCD'))
len(permute('ABCD'))
```
